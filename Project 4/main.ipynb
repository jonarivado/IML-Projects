```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso, Ridge, LassoCV, RidgeCV
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
df_train=pd.read_csv(r'C:\Users\Manue\Documents\GitHub\IML-Projects\Project 4\data\train_features.csv')
df_labels=pd.read_csv(r'C:\Users\Manue\Documents\GitHub\IML-Projects\Project 4\data\train_labels.csv')
df_test=pd.read_csv(r'C:\Users\Manue\Documents\GitHub\IML-Projects\Project 4\data\test_features.csv')
display(df_test)
X=df_train.drop(columns=['Id', 'smiles'])
y=df_labels.drop(columns=['Id'])
test_set=df_test.drop(columns=['Id', 'smiles'])
display(X,y,test_set)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso, Ridge, LassoCV, RidgeCV
from sklearn.metrics import mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(X, np.ravel(y), test_size=0.2, shuffle=True, random_state=3)

names = [
    "Lasso",
    "Ridge",
    "LassoCV",
    "RidgeCV",
]

models = [
    Lasso(),
    Ridge(),
    LassoCV(),
    RidgeCV(),
]

# iterate over linear models
for name, model_ in zip(names, models):
    model_.fit(X_train, y_train)
    y_pred=model_.predict(X_test)
    err=np.sqrt(mean_squared_error(y_test, y_pred))
    display(name, err)
    
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, np.ravel(y), test_size=0.2, shuffle=True, random_state=3)
model=Ridge()
model.fit(X_train, y_train)
y_pred=model.predict(X_test)
err=np.sqrt(mean_squared_error(y_test, y_pred))
display(err)
test_pred=model.predict(test_set)
```

```python
submission_df=pd.read_csv(r'C:\Users\Manue\Documents\GitHub\IML-Projects\Project 4\data\sample.csv')
submission_df['y']=test_pred
submission_df.to_csv('predictions.csv', index= False, header=True)
```
