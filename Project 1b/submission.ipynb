```python
from sklearn.metrics import mean_squared_error
import pandas as pd
from sklearn.linear_model import RidgeCV, LassoCV, LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
trainset = pd.read_csv('train.csv')
X = trainset.drop(trainset.columns[[0,1]],axis=1)
y = pd.DataFrame(trainset['y'])
```

```python
#feature transformation
#quadratic x6-x10
X_squared = X**2
X_squared.columns=['x6', 'x7','x8', 'x9', 'x10']
X_trafo = pd.concat([X, X_squared],axis=1)

#exponential x11-x15
X_exp = np.exp(X)
X_exp.columns=['x11', 'x12','x13', 'x14', 'x15']
X_trafo = pd.concat([X_trafo, X_exp],axis=1)

#cosine x16-x20
X_cos = np.cos(X)
X_cos.columns=['x16', 'x17','x18', 'x19', 'x20']
X_trafo = pd.concat([X_trafo, X_cos],axis=1)

#constant x21
X_trafo['x21'] = np.ones_like(700)

X_train, X_test, y_train, y_test = train_test_split(
    X_trafo, y, test_size=0.05, random_state=0)

display(X_trafo)
```

```python
reg = LassoCV(random_state=42069, max_iter=1000000,cv=75, tol=1e-6).fit(X_train, np.ravel(y_train))

y_pred = reg.predict(X_test)
```

```python
display(mean_squared_error(y_test,y_pred)**0.5)
```

```python
display(reg.coef_)
pd.DataFrame(reg.coef_).to_csv('submission.csv', index=False, header=False)
```
